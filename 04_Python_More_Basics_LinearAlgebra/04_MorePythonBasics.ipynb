# Debugging

```python
def factorial(x):
    if x == 0 or x == 1:
        return 1
    return x * factorial(x-1)
```

```python
factorial(5)
```

```python
import random
```

```python
def code_to_debugg():
    import pdb; pdb.set_trace() # python package
    for i in range(10):
        x = random.random()
        factorial(x)
```

```python
code_to_debugg()
```

```python
code_to_debugg()
```
